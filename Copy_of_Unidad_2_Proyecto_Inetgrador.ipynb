```python
#Deportistas cuyo ranking ha subido al menos dos lugares entre 2010 y 2020.
df = pd.read_csv('forbes-dataset-2020.csv')
df
```

```python
df1 = df.dropna()
df1
```

```python
df1.to_csv("tabla_completa.csv", index = False)
from google.colab import files
files.download("tabla_completa.csv")
```

```python
df_ej_1 = df1[df.Year >= 2010]
df_ej_1
```

```python
diccionario = {ord(">"):None,ord("?"):None}
df_ej_1["Previous Year Rank"] = df1["Previous Year Rank"].str.translate(diccionario)
df_ej_1
```

```python
df_ej_1 = df_ej_1.drop (labels=[211,225,226,286,287], axis=0)
df_ej_1
```

```python
df_ej_1["Previous Year Rank"] = df_ej_1["Previous Year Rank"].astype(int)
```

```python
df_ej_1 ["Jugadores que subieron de rango"] = np.where((df_ej_1 ["Previous Year Rank"] > df_ej_1 ["Current Rank"]),(df_ej_1 ["Previous Year Rank"] - df_ej_1 ["Current Rank"]),np.nan)
df_ej_1
```

```python
df_ej_1_2 = df_ej_1.dropna()
df_ej_1_2 = df_ej_1_2.drop(["S.NO"],axis =1)
df_ej_1_2
```

```python
df_ej_1_2.to_csv('Ejercicio_1.csv', index = False)
```

```python
#Deporte y país con mayor número de atletas no rankeados que entraron en la lista de atletas mejor pagados.

table = pd.pivot_table (df1, values = "earnings ($ million)", index = ["Sport","Name"], columns= ["Nationality"],aggfunc=np.sum, fill_value=0)
table
```

```python
# 4. País con mayor número de deportes en el Dataset
pais_top = df1['Nationality'].value_counts().to_frame()
pais_top.style.background_gradient(cmap='Greens') 
```

```python
# 5. ¿Cuántos atletas por deporte tiene cada país?
df_no_duplicados = df1.drop_duplicates(subset = ['Name'])
df_ej_5 = pd.DataFrame(df_no_duplicados.groupby(['Nationality', 'Sport'])['Name'].count())
atletas_por_deporte_pais = df_ej_5.rename(columns = {'Name': 'Número de atletas'})
atletas_por_deporte_pais.style.background_gradient(cmap='Reds') 
```
