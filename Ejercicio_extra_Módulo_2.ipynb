```python
import pandas as pd
import numpy as np
import requests
```

```python
df= pd.read_csv('data.csv')
df.columns
```

```python
#Hacemos una pequeña limpieza del DF
#Renombramos las columnas por nombres más amigables para el tecleo xd
#Los datos vacíos por el momento los remplazamos por 0s
df= df.rename(columns={'Escribe un alias de al menos 3 caracteres (Sólo tú sabrás quién eres)':'Alias','¿Cuál es tu género?':'Genero','Cuál es tu rango de edad':'Rango_Edad','¿Qué porcentaje del tiempo te consideras introvertido/a? [Intro]':'Porcentaje_Intro','¿Qué porcentaje del tiempo te consideras introvertido/a? [Extro]':'Porcentaje_Extro','¿Cuántas horas duermes en promedio en una semana?':'Horas_durm','¿A qué hora te acuestas en promedio en una semana?':'Hora_durmicion','¿Cuántos países has visitado?':'Paises_visit','En la escala del 1 al 5 ¿Qué nivel de inseguridad sientes al vivir en tu país de nacimiento? (1 muy seguro - 5 muy inseguro)':'Escala_ins_pais','¿En qué estado (geográfico) has vivido la mayor parte de tu vida?':'Estado_geo','¿Cuántas frutas comes en un día?':'Num_frutas','Escoge uno':'Animales','Nombra tus 3 shows, series, películas o animes favoritos separados por comas (Ej. Harry Potter, Game of Thrones, Pokémon)':'Shows_fav'})
df= df.fillna(0)
df.columns
df['Shows_fav']=df['Shows_fav'].astype(str)
df['Shows_fav']=df['Shows_fav'].map(lambda x: x.replace(' y ',','))
```

```python
#Limpieza de columna Horas_durm
df['Horas_durm']=df['Horas_durm'].astype(str)
df['Horas_durm']=df['Horas_durm'].str.extract((r'(\d+)'))
df['Horas_durm']=df['Horas_durm'].astype(int)
```

```python
#Limpieza de columna Paises_visit
df['Paises_visit']=df['Paises_visit'].astype(str)
df['Paises_visit']=df['Paises_visit'].str.extract((r'(\d+)'))
df['Paises_visit']=df['Paises_visit'].astype(int)
```

```python
#Limpieza de columna Estado_geo 
df['Estado_geo']=df['Estado_geo'].str.upper()
for i in range(len(df['Estado_geo'])):
  df['Estado_geo'][i]=df['Estado_geo'][i].strip()

df['Estado_geo']=df['Estado_geo'].str.replace('CIUDAD DE MÉXICO','CDMX').replace('EDOMEX','ESTADO DE MÉXICO')
```

```python
#Eliminar datos extra en la columna Porcentaje Intro y extro
df['Porcentaje_Intro']=df['Porcentaje_Intro'].astype(str)
df['Porcentaje_Intro']=df['Porcentaje_Intro'].map(lambda x: x.replace('%',''))
df['Porcentaje_Intro']=df['Porcentaje_Intro'].astype(int)
```

```python
df['Porcentaje_Extro']=df['Porcentaje_Extro'].astype(str)
df['Porcentaje_Extro']=df['Porcentaje_Extro'].map(lambda x: x.replace('%',''))
df['Porcentaje_Extro']=df['Porcentaje_Extro'].astype(int)
```

```python
#Rellenar los datos faltantes de cada porcentaje
for i in range(len(df['Porcentaje_Intro'])):
   if df['Porcentaje_Intro'][i] == 0:
     df['Porcentaje_Intro'][i]=(100-df['Porcentaje_Extro'][i])
   else:
     pass
for i in range(len(df['Porcentaje_Extro'])):
   if df['Porcentaje_Extro'][i] == 0:
     df['Porcentaje_Extro'][i]=(100-df['Porcentaje_Intro'][i])
   else:
     pass
```

```python
#Hacer una columnna por cada show
#eliminamos columnas innecesarias
shows=df['Shows_fav'].str.split(',',expand=True)
shows.columns=['Show1','Show2', 'Show3','nulo']

df= pd.concat([df, shows], axis=1)
df.drop(columns=['Marca temporal','nulo','Shows_fav','¿Qué otra pregunta te gustaría incluir?'], inplace=True)
```

```python
#Correción de errores humanos
df=df.replace('Game of throne','Game of thrones').replace('How i metí Your mother','How i met Your mother').replace('How I Meet your Mother','How i met Your mother').replace('HIMYM','How i met Your mother').replace(' AoT','Shingeki No Kiojin').replace('Theory big bang','The Big Bang Theory').replace(' Naruto.','Naruto').replace('Full Metal Alchemist Brotherhood','Full Metal Alchemist').replace('GOT','Game of thrones').replace('Pokemon','Pokémon').replace('Lord of the Rings','The Lord of the rings')
```

```python
#Pasamos todo a mayúsculas y eliminamos espacios en blanco
df_=df.copy()
df_['Show1']=df_['Show1'].astype(str)
df_['Show2']=df_['Show2'].astype(str)
df_['Show3']=df_['Show3'].astype(str)
df_['Show1']=df_['Show1'].str.upper().map(lambda x: x.strip())
df_['Show2']=df_['Show2'].str.upper().map(lambda x: x.strip())
df_['Show3']=df_['Show3'].str.upper().map(lambda x: x.strip())
```

```python
df_.insert(15, "Puntos", 0, allow_duplicates=False) #en la posicion 15, llamada puntos, donde todos los valores son 0
df_["Puntos"]=df_["Puntos"].astype(int)
```

```python
#Creamos una función llamada "Tu_match_perfecto" para encontrar a la persona más parecida a ti en el cuestionario 
def Tu_match_perfecto(Alias):
  #Encontramos al alias en el DF y generamos un DF de él.
  persona = df_.loc[df_['Alias']==Alias]
  #columnas que queremos eliminar 
  columnas=['Genero', 'Rango_Edad', 'Porcentaje_Intro', 'Porcentaje_Extro',
       'Horas_durm', 'Hora_durmicion', 'Paises_visit', 'Escala_ins_pais',
       'Estado_geo', 'Num_frutas', 'Animales', 'Show1', 'Show2', 'Show3']
  #Se crea un DF por columna a separar y se le suma 1 
  P1=df_.loc[df_["Genero"]==persona["Genero"].item()]
  P1.drop(columns=columnas,inplace=True)
  P1["Puntos"]=+1
  P2=df_.loc[df_["Rango_Edad"]==persona["Rango_Edad"].item()]
  P2.drop(columns=columnas,inplace=True)
  P2["Puntos"]=+1
  P3=df_.loc[df_["Porcentaje_Intro"]==persona["Porcentaje_Intro"].item()]
  P3.drop(columns=columnas,inplace=True)
  P3["Puntos"]=+1
  P4=df_.loc[df_["Horas_durm"]==persona["Horas_durm"].item()]
  P4.drop(columns=columnas,inplace=True)
  P4["Puntos"]=+1
  P5=df_.loc[df_["Paises_visit"]==persona["Paises_visit"].item()]
  P5.drop(columns=columnas,inplace=True)
  P5["Puntos"]=+1
  P6=df_.loc[df_["Escala_ins_pais"]==persona["Escala_ins_pais"].item()]
  P6.drop(columns=columnas,inplace=True)
  P6["Puntos"]=+1
  P7=df_.loc[df_["Estado_geo"]==persona["Estado_geo"].item()]
  P7.drop(columns=columnas,inplace=True)
  P7["Puntos"]=+1
  P8=df_.loc[df_["Num_frutas"]==persona["Num_frutas"].item()]
  P8.drop(columns=columnas,inplace=True)
  P8["Puntos"]=+1
  P9=df_.loc[df_["Animales"]==persona["Animales"].item()]
  P9.drop(columns=columnas,inplace=True)
  P9["Puntos"]=+1
  P10=df_.loc[df_["Porcentaje_Extro"]==persona["Porcentaje_Extro"].item()]
  P10.drop(columns=columnas,inplace=True)
  P10["Puntos"]=+1
  #se crea nuevo DF y se mergean los anteriores para al final proceder a sumar los puntos y eliminar las columnas inecesarias
  PT=pd.merge(P1, P2,
  how='outer', on='Alias')
  PT=pd.merge(PT, P3,
  how='outer', on='Alias')
  PT=pd.merge(PT, P4,
  how='outer', on='Alias')
  PT=pd.merge(PT, P5,
  how='outer', on='Alias')
  PT=pd.merge(PT, P6,
  how='outer', on='Alias')
  PT=pd.merge(PT, P7,
  how='outer', on='Alias')
  PT=pd.merge(PT, P8,
  how='outer', on='Alias')
  PT=pd.merge(PT, P9,
  how='outer', on='Alias')
  PT=pd.merge(PT, P10,
  how='outer', on='Alias')
  PT=PT.fillna(0)
  PT['Puntos_total'] = PT.sum(axis = 1)
  PT.drop(columns=['Puntos_x', 'Puntos_y', 'Puntos_x', 'Puntos_y', 'Puntos_x','Puntos_y', 'Puntos_x', 'Puntos_y', 'Puntos_x', 'Puntos_y'], inplace=True)
  #se crea DF por show a comparar y se le suma 1 punto
  S1=df_.loc[(df_["Show1"]==persona["Show1"].item())|(df_["Show1"]==persona["Show2"].item())|(df_["Show1"]==persona["Show3"].item())]
  S1.drop(columns=columnas,inplace=True)
  S1["Puntos"]=+1
  S1

  S2=df_.loc[(df_["Show2"]==persona["Show1"].item())|(df_["Show2"]==persona["Show2"].item())|(df_["Show2"]==persona["Show3"].item())]
  S2.drop(columns=columnas,inplace=True)
  S2["Puntos"]=+1
  S2

  S3=df_.loc[(df_["Show3"]==persona["Show1"].item())|(df_["Show3"]==persona["Show2"].item())|(df_["Show3"]==persona["Show3"].item())]
  S3.drop(columns=columnas,inplace=True)
  S3["Puntos"]=+1
  S3

  #se crea nuevo DF y se mergean los anteriores para al final proceder a sumar los puntos y eliminar las columnas inecesarias
  ST=pd.merge(S1, S2,
  how='outer', on='Alias')
  ST=pd.merge(ST, S3,
  how='outer', on='Alias')
  ST=pd.merge(ST, df_,
  how='outer', on='Alias')
  ST.drop(columns=columnas,inplace=True)

  ST=ST.fillna(0)
  ST['Puntos_total_S'] = ST.sum(axis = 1)

  ST.drop(columns=['Puntos_x', 'Puntos_y', 'Puntos_x', 'Puntos_y', 'Puntos_x','Puntos_y', 'Puntos_x', 'Puntos_y'], inplace=True)
  #se mergean los DF con las sumas y se vuelven a sumar y eliminar columnas inecesarias, se acomoda de mayor a menor
  PT=pd.merge(PT, ST,
  how='outer', on='Alias')
  PT=PT.fillna(0)
  PT['Porcentaje de compatibilidad'] = PT.sum(axis = 1)
  PT.drop(columns=['Puntos_total', 'Puntos_total_S'], inplace=True)
  PT.sort_values('Porcentaje de compatibilidad', ascending=False,inplace=True)
  #se elimina la fila del usuario que se está buscando
  PT=PT.loc[PT['Porcentaje de compatibilidad']<13]
  #se combierte a porcentaje el resultado
  PT['Porcentaje de compatibilidad']=((PT['Porcentaje de compatibilidad']/13)*100).round(2).astype(str) + '%'

  return PT 


```
